```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
total = 0
accepted = 0
cooling = 0.95 ** 50


for _ in range(1000000):
    total += 1
    x = np.random.randn()
    if x < 0:
        accepted += 1
    else:
        prob = np.exp(-x / cooling)
        if np.random.uniform() < prob:
            accepted += 1

print(accepted / total - 0.5)
```

# Cooling schedule

```python
fig, ax = plt.subplots(figsize=(10, 8))
T = 0.2
T2 = 0.0011841058440667995
Tfactor = -np.log(T/T2)
rate = 0.95
steps = 100
X = []
y = []
y2 = []
for i in range(steps+1):
    X.append(i)
    y.append(T*rate**i)
    y2.append(T * np.exp(Tfactor * i / steps))
plt.plot(X, y)
plt.plot(X, y2)
print(y[-1])
```

```python
for x1, x2 in zip(y, y2):
    print(x1 - x2)
```
