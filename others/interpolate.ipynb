```python
from scipy import interpolate
import numpy as np
```

```python
x = np.arange(1, 11, 1)
y = np.arange(1, 11, 1)
```

```python
x, y
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
xx
```

```python
yy
```

```python
z = xx + yy
```

```python
z
```

```python
f = interpolate.interp2d(x, y, z)
```

```python
xnew = np.arange(1, 11, 0.5)
ynew = np.arange(1, 11, 0.5)
```

```python
f(xnew, ynew)
```
