```python
import pandas
import numpy as np
import matplotlib.pyplot as plt
```

# Overall Average

```python
df = pandas.read_csv('ReviewScores.csv')
df['mean'] = df.mean(axis=1)
df
```

```python
df.sort_values('mean', ascending=False)
```

# Reviewer Analysis

```python
columns = df.columns
columns = list(columns)[:-1]
bins = np.linspace(0.5, 10.5, 11)
```

```python
for col in columns:
    plt.hist(df[col].values, bins=bins, edgecolor='k')
    plt.ylim([0, 16])
    plt.title(f'Reviewer {col}')
    plt.show()
```
