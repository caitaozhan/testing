```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.size'] = 25
```

```python
mean1   = np.array([20, 30, 40, 25, 35])
std1    = np.array([1, 2, 3, 2, 1])
mean2   = np.array([15, 25, 35, 30, 40])
std2    = np.array([1, 2, 1.5, 2, 1])
mean3   = np.array([10, 15, 20, 15, 18])
std3    = np.array([1, 2, 1.5, 2, 1])
```

```python
ind   = np.arange(len(mean1))
width = 0.25
```

```python
fig, ax = plt.subplots(figsize=(12, 9))
ax.bar(ind-width, mean1, width, yerr=std1, label='Our')
ax.bar(ind,       mean2, width, yerr=std2, label='SPLOT')
ax.bar(ind+width, mean3, width, yerr=std3, label='cluster')
ax.set_title('Errors by Algorithms and # of Intruder', fontsize=30)
ax.legend()
ax.set_ylabel('Error (m)')
#ax.set_yticks(fontsize=20)
ax.set_xticks(ind)
ax.set_xticklabels(('1', '3', '5', '7', '10'))
ax.set_xlabel('Varing # of Intruder')
fig.savefig('error.png')
```
