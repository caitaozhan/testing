```python
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
plt.rcParams['font.size'] = 60
plt.rcParams['lines.linewidth'] = 10
plt.rcParams['lines.markersize'] = 15
```

```python
fig, ax = plt.subplots(figsize=(16, 16))
fig.subplots_adjust(left=0.2, right=0.96, top=0.8, bottom=0.15)

octopus = np.loadtxt('octopus.txt')
d = defaultdict(list)
for n, t in octopus:
    d[n].append(t)
X, Y = [], []
for key, val in d.items():
    X.append(int(key))
    Y.append(np.mean(val)*1000)
plt.plot(X, Y, marker='o', label='Optimal')

octopus = np.loadtxt('octopus-greedy.txt')
d = defaultdict(list)
for n, t in octopus:
    d[n].append(t)
X, Y = [], []
for key, val in d.items():
    X.append(int(key))
    Y.append(np.mean(val)*1000)
plt.plot(X, Y, marker='o', label='Greedy')
ax.set_xlabel('Number of Nodes')
ax.set_xlim([0, 1050])
ax.set_xticks([100, 250, 500, 750, 1000])
ax.set_ylabel('Time (milliseconds)')
ax.set_yticks(range(8))
box_to_anchor = (-0.04, 1.04)
ax.legend(bbox_to_anchor=box_to_anchor, loc='lower left', ncol=2, fontsize=50)
ax.tick_params(direction='in', length=15, width=5)
fig.savefig('runtime.png')
```
