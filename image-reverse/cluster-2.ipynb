```python
from PIL import Image
import numpy as np
```

## Cluster 2: 456 racks

```python
im = Image.open('cluster-2.png').convert('L')  # RGB to gray scale
im
```

```python
im.mode, im.size
```

```python
left   = 107
top    = 22
right  = 348
bottom = 206
im = im.crop((left, top, right, bottom))
im
```

```python
im = im.resize((456, 456))   # there are approximately 456 nodes
im
```

```python
im_array = np.array(im.getdata()).reshape((456, 456))
im_array[10][::20]
```

```python
im_array = (255-im_array)/255 # scale to [0, 1]
im_array = im_array*8         # scale to [0, 8]
im_array[10][::20]
```

```python
im_array = np.power(10., im_array)  # scale to [0, 1e8]
im_array[10][::20]
```

```python
im_array = np.where(im_array<=1, 0, im_array)  # the 1 should be 0 traffic
im_array[10][::20]
```

```python
np.savetxt('cluster-2.txt', im_array, delimiter=',')
```
