```python
from PIL import Image
import numpy as np
```

## Cluster 3:  1412 racks

```python
im = Image.open('cluster-3.png').convert('L')  # RGB to gray scale
im
```

```python
im.mode, im.size
```

```python
left   = 126
top    = 22
right  = 352
bottom = 206
im = im.crop((left, top, right, bottom))
im
```

```python
im = im.resize((1412, 1412))   # there are approximately 1412 nodes
im
```

```python
im_array = np.array(im.getdata()).reshape((1412, 1412))
im_array[30][::40]
```

```python
im_array = (255-im_array)/255 # scale to [0, 1]
im_array = im_array*8         # scale to [0, 8]
im_array[30][::40]
```

```python
im_array = np.power(10., im_array)  # scale to [0, 1e8]
im_array[30][::40]
```

```python
im_array = np.where(im_array<=1, 0, im_array)  # the 1 should be 0 traffic
im_array[30][::40]
```

```python
np.savetxt('cluster-3.txt', im_array, delimiter=',')
```
