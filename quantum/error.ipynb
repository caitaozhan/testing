```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def error(t: float, T: float) -> float:
    """
    Args:
        t: the time
        T: relaxing time
    Return:
        the error probability
    """
    return 1 - np.exp(-t/T)
```

```python
def error2(t: float, T1: float, T2: float) -> float:
    """
    Args:
        t: the time
        T1: relaxation time
        T2: dephasing time
    Return:
        the error probability
    """
    return 3/4 - 0.25*np.exp(-t/T1) - 0.5*np.exp(-t/T2)
```

```python
T1 = 1

X = np.linspace(0, 2, 201)
y = [error(x, T1) for x in X]

fig, ax = plt.subplots(figsize=(8, 4))
plt.plot(X, y)
ax.set_xlim([0, 2])
ax.set_ylim([0, 1])
ax.set_xlabel('Time')
ax.set_ylabel('Error')
ax.grid()
```

```python
T1 = 3
T2 = 1

X = np.linspace(0, 2, 201)
y = [error2(x, T1, T2) for x in X]

fig, ax = plt.subplots(figsize=(8, 4))
plt.plot(X, y)
ax.set_xlim([0, 2])
ax.set_ylim([0, 1])
ax.set_xlabel('Time')
ax.set_ylabel('Error')
ax.grid()
```
