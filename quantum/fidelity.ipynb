```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# two EP has the same fidelity
# Dur and Briegel (2007) formula (18) page 14
def f(F):
    return (F**2 + ((1-F)/3)**2) / (F**2 + 2*F*(1-F)/3 + 5*((1-F)/3)**2)
```

```python
# two EP have different fidelities
def f2(x, y):
    return (x*y + ((1-x)*(1-y)/9)) / (x*y + x*(1-x)/3 + y*(1-y)/3 + 5*(1-x)*(1-y)/9)
```

## Two EP are the same

```python
X = np.linspace(0, 1, 101)
y = []
for x in X:
    y.append(f(x))
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

increase = [a-b for a, b in zip(y, X)]
print('max increase =', max(increase[50:]))
ax.plot(X, y, label="pufication")
ax.plot(X, X, label="no purification")
ax.plot(X, increase, label="fidelity increase")
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.grid()
ax.legend()
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

increase = [a-b for a, b in zip(y, X)]
print('max increase =', max(increase[50:]))
ax.plot(X[50:], increase[50:], label="fidelity increase")
ax.set_xlim([0.5, 1])
ax.set_ylim([0, 0.04])
ax.grid()
ax.legend()
```

## Two EP are different

```python
y2 = []
for x in X:
    y2.append(f2(x, x-0.04))
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

increase = [a-b for a, b in zip(y2, X)]
lower = [x - 0.04 for x in X]
print('max increase =', max(increase[50:]))
ax.plot(X, y2, label="pufication")
ax.plot(X, X, label="no purification")
ax.plot(X, increase, label="fidelity increase")
# ax.plot(X, lower, label="x - 0.04")
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.grid()
ax.legend()
```
