# Visualize all the Layers in https://github.com/mit-han-lab/torchquantum/blob/master/torchquantum/layers.py

```python
import torchquantum as tq
from torchquantum.plugins.qiskit_plugin import tq2qiskit
```

# Random Layer (only RX, RY, RZ, CNOT)

```python
# np.random.seed(42)
# torch.manual_seed(42)
n_qubits = 4
dev = tq.QuantumDevice(n_wires=n_qubits)
rlayer = tq.RandomLayer(n_ops=16, wires=list(range(n_qubits)))
circ = tq2qiskit(dev, rlayer)
circ.draw('mpl')
```

# Random Layer (all types of gates, some connot be converted to Qiskit)

```python
n_qubits = 4
dev = tq.QuantumDevice(n_wires=n_qubits)
rlayer = tq.RandomLayerAllTypes(n_ops=16, wires=list(range(n_qubits)))
circ = tq2qiskit(dev, rlayer)
circ.draw('mpl')
```

```python
n_qubits = 3
dev = tq.QuantumDevice(n_wires=n_qubits)
qlayer = tq.layers.SimpleQLayer(n_wires=n_qubits)
circ = tq2qiskit(dev, qlayer)
circ.draw('mpl')
```
